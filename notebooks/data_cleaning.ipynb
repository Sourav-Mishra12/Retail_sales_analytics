```python
import pandas as pd
import chardet
```

```python
file_path = "../data/sales_data_sample.csv"

with open(file_path, "rb") as f:
    result = chardet.detect(f.read(100000))
    print(result)


df = pd.read_csv(file_path , encoding='Windows-1252')
df.head()
df.columns
# df.isnull().sum()
# (df.isnull().sum() / len(df)) * 100
```

```python
df['ADDRESSLINE2'].fillna(' ' , inplace=True)
df['STATE'].fillna('unknown', inplace=True)
df['TERRITORY'].fillna('not assigned', inplace=True)
df['POSTALCODE'].fillna('unknown', inplace=True)

# df.isnull().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['REVENUE'] = df['QUANTITYORDERED'] * df['PRICEEACH']

df.rename(columns={
    "QTR_ID" : "QUARTER",
    "MONTH_ID" : "MONTH" , 
    "YEAR_ID" : "YEAR"
} , inplace=True)

df = df[df['STATUS'].isin(['Shipped','Resolved','Cancelled','On Hold'])]

df.head()
```

```python
df.to_csv("clean_sales_data.csv",index=False)
print("data saved successfully")
df.to_excel("clean_data.xlsx" , index=False)
```
